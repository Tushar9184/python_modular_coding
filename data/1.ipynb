```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"],format="%d-%m-%Y") 
```

```python
df["day"]=df["Order_Date"].dt.day
df["month"]=df["Order_Date"].dt.month
df["year"]=df["Order_Date"].dt.year 
df.drop("Order_Date",axis=1,inplace=True)
```

```python
df.dropna(subset="Time_Orderd",inplace=True)
```

```python
def extract_time(x):
    try:
        return x.split(":")[0]+':'+x.split(':')[1][:2]
    except IndexError:
        return "00:00"
```

```python
df["Time_Orderd"] = pd.to_datetime(df["Time_Orderd"], format="%H:%M", errors="coerce")

df["Timeorder_hour"] = df["Time_Orderd"].dt.hour
df["Timeorder_minute"] = df["Time_Orderd"].dt.minute

# Convert to int safely (handles NaN)
df["Timeorder_hour"] = df["Timeorder_hour"].fillna(0).astype(int)
df["Timeorder_minute"] = df["Timeorder_minute"].fillna(0).astype(int)
```
